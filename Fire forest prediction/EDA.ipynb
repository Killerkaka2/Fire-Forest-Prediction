# Data Cleaning

### missing values

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

The dataset is in two parts/regions : <br>
1] Bejaia region - from index 0 to 121  (given number = 0)<br>
2] Sidi-Bel Abbes Region - from index 123 onwards (given number = 1)<br><br>
Therefore make a new column with regions embedded in it

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
```

```python
dataset.info()
```

```python
dataset['Region'] = dataset["Region"].astype(int)
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.dropna().reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
dataset = dataset.drop(122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
dataset.columns
```

## <br><hr><br>

### To remove extra spaces in column/label names

```python
dataset.columns = dataset.columns.str.strip()
dataset.columns
```

## <hr>

## To change the datatypes of first 4 columns as int 

```python
dataset.columns
```

```python
dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
dataset.info()
```

### To change objects datatypes to float USING LOGIC

```python
objects = [features for features in dataset.columns if dataset[features].dtypes == 'O']
```

```python
objects
```

```python
for i in objects:
    if i!= "Classes":
        dataset[i] = dataset[i].astype(float)
```

```python
dataset.info()
```

```python
dataset
```

```python
dataset.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
